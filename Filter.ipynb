```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
import math
```

```python
#img = imread('material/lena.png', as_gray=True)
img = imread('material/lena.png')
print('shape:', img.shape)
plt.imshow(img,cmap=plt.cm.gray)
plt.show()
```

## Gaussian Filter

```python
?ndimage.gaussian_filter()
```

```python
gau = ndimage.gaussian_filter(img, sigma=(1/(2*math.log(2))))
plt.imshow(gau.astype(np.float32),cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

## Edge Detect
https://en.wikipedia.org/wiki/Sobel_operator

```python
# Get x-gradient in "sx"
print('sx')
sx = ndimage.sobel(img,axis=0,mode='constant')
plt.imshow(sx.astype(np.float32),cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# Get y-gradient in "sy"
print('sy')
sy = ndimage.sobel(img,axis=1,mode='constant')
plt.imshow(sy.astype(np.float32),cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# Get square root of sum of squares
print('sob')
sob=np.hypot(sx,sy)
# Hopefully see some edges
plt.imshow(sob.astype(np.float32),cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
