```python
import numpy as np
from scipy import ndimage, misc
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
import math
```

```python
#img = imread('material/lena.png', as_gray=True)
img = imread('material/lena.png', as_gray  = True)
print('shape:', img.shape)
plt.gray()
plt.imshow(img)
plt.show()
```

## Gaussian Filter

```python
s = (1/(2*math.log(2)))
s = 5
print('sigma=', s)
gau = ndimage.gaussian_filter(img, sigma=s)
plt.imshow(gau,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

## Edge Detect
https://en.wikipedia.org/wiki/Sobel_operator

```python
# Get x-gradient in "sx"
print('sx')
sx = ndimage.sobel(img,axis=0,mode='constant')
print(sx.min())
plt.imshow(sx,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# Get y-gradient in "sy"
print('sy')
sy = ndimage.sobel(img,axis=1,mode='constant')
plt.imshow(sy,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# Get square root of sum of squares
print('sob')
sob=np.hypot(sx,sy)
# Hopefully see some edges
plt.imshow(sob.astype(np.int16),cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
